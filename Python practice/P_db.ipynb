```python
class Database:
    def __init__(self):
        self.__student_records = []
        self.__employee_records = []
        self.__finance_records = []

    def add_student(self, student_id, name, age, grade, major, gpa, email, phone):
        student = {
            'student_id': student_id,
            'name': name,
            'age': age,
            'grade': grade,
            'major': major,
            'gpa': gpa,
            'email': email,
            'phone': phone
        }
        self.__student_records.append(student)

    def get_student(self, student_id):
        for student in self.__student_records:
            if student['student_id'] == student_id:
                return student
        return None

    def get_all_students(self):
        return self.__student_records

    def add_employee(self, employee_id, name, position, salary):
        employee = {'employee_id': employee_id, 'name': name, 'position': position, 'salary': salary}
        self.__employee_records.append(employee)

    def get_employee(self, employee_id):
        for employee in self.__employee_records:
            if employee['employee_id'] == employee_id:
                return employee
        return None

    def get_all_employees(self):
        return self.__employee_records

    def add_finance_record(self, record_id, description, amount, date):
        finance_record = {'record_id': record_id, 'description': description, 'amount': amount, 'date': date}
        self.__finance_records.append(finance_record)

    def get_finance_record(self, record_id):
        for record in self.__finance_records:
            if record['record_id'] == record_id:
                return record
        return None

    def get_all_finance_records(self):
        return self.__finance_records

    def set_student_records(self, records):
        self.__student_records = records

    def set_employee_records(self, records):
        self.__employee_records = records

    def set_finance_records(self, records):
        self.__finance_records = records

    def get_student_records(self):
        return self.__student_records

    def get_employee_records(self):
        return self.__employee_records

    def get_finance_records(self):
        return self.__finance_records


db = Database()

db.add_student(1, "John Doe", 20, "A", "Computer Science", 3.8, "john.doe@example.com", "555-1234")
db.add_employee(101, "Jane Smith", "Manager", 75000)
db.add_finance_record(201, "Office Supplies", 500, "2024-08-25")


print("All Students:", db.get_all_students())
print("All Employees:", db.get_all_employees())
print("All Finance Records:", db.get_all_finance_records())

print("Student with ID 1:", db.get_student(1))
print("Employee with ID 101:", db.get_employee(101))
print("Finance Record with ID 201:", db.get_finance_record(201))
```
